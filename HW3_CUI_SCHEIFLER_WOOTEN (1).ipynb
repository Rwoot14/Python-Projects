Q3

We first needed to calculate the number of turks who correctly chose the category. After getting those value counts, we then calculated the number of ratings each turk completed. After getting the total number of ratings, we divided the number of correct ratings, by the number of total ratings for each turk and called that average_corr. Then we made a new data frame with the number of ratings, and the average correctness called rater_goodness.

```python
g=labels_on_gold.loc[labels_on_gold['cat_x']==labels_on_gold['cat_y']]
correct=g['turk'].value_counts()
correct_frame=pd.DataFrame(g['turk'].value_counts())
```

```python
num_ratings=pd.DataFrame(labels_on_gold['turk'].value_counts())
```

```python
rater_goodness=pd.DataFrame(correct_frame['turk']/num_ratings['turk'])
rater_goodness['total']=labels_on_gold['turk'].value_counts()
rater_goodness=rater_goodness.rename(columns={'turk':'average_corr','total':'ratings'})
rater_goodness=rater_goodness.fillna(value=0)
rater_goodness
```

Q4

We added a new collumn to rater_goodness called odds which was average_corr/(1.001-average_corr)

```python
rater_goodness['odds']=rater_goodness['average_corr']/(1.001-rater_goodness['average_corr'])
rater_goodness[:10]
```

Q7

We first found the number of ratings a turk needed for being in the top 75% of all turks. Then we created a mask using that cutoff and created a dataframe with only the top 75% of turks (in terms of rating amounts). Then we merged the the labels unkown data frame, with the rater_75 data frame, and conducted a groupby to get the product of turk odds for each url's category. 

```python
rater_goodness.quantile(q=.75)
```

```python
rater_75=rater_goodness[rater_goodness['ratings']>=11]
rater_75['turk']=rater_75.index
rater_75
```

```python
labels_unknown_rater_75=labels_unknown.merge(rater_75,on='turk')
overall_odds=labels_unknown_rater_75.groupby(['url','cat'])[['odds']].prod()
overall_odds


```

Q8

We first unstacked the overall odds data frame and found the max odds for each url's category and created a new data frame which listed a url's top odds and the corresponding category. 

```python
b=overall_odds.unstack(-1)
b=b['odds']
b
```

```python
b['top odds']=b[['G','P','R','X']].max(axis=1)

b['top category']=b[['G','P','R','X']].idxmax(axis=1)
result_75=DataFrame([b['top odds'],b['top category']])
result_75=result_75.transpose()
result_75
```
